### 13. Написати функцію, яка б множила всі значення в колонці, які менше середнього значения в цій колонці, на 2, і ділила інші значення на 4.

```python
def myfunc(x):
    xx=x.copy()
    if x.ndim>1:
        for i in range(x.shape[0]):
            xx[np.where(x[:,i]<(x[:,i].mean())),i] *=2
            xx[np.where(x[:,i]>=(x[:,i].mean())),i] *=0.25
    elif x.ndim==1:
            xx[np.where(x<(x.mean()))] *=2
            xx[np.where(x>=(x.mean()))] *=0.25
    return xx
```

```python
x = np.array([[4,1,1], [1,0,4], [1,1,2]]).astype('float32')
print('Input array: ','\n', x)
print('Processed array: ', '\n', myfunc(x))
```

### 14. Застосувати отриману функцію до 3-ї колонки

```python
x = np.array([[4,1,1], [1,0,4], [1,1,2]]).astype('float32')

print(myfunc(x[:,2]))
```
